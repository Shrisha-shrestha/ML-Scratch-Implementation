```python
# Import required libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Import google drive
from google.colab import drive
drive.mount('/content/drive')

# Update the file path based on the location of file.
file_path = "/content/drive/MyDrive/Colab Notebooks/Applied Machine Learning/Linear Regression/data.csv"
df = pd.read_csv(file_path, header=None)
df = df.dropna()
```

```python
# Define Scaling function to scale out the data
def scale(d):
    mean=d.mean()
    sd=d.std()

    return (d-mean)/sd
```

```python
# Split out the features and labels from the dataframe and scale them
features = scale(np.reshape(np.array(df.iloc[:, 0]), (len(df), 1)))
labels = scale(np.reshape(np.array(df.iloc[:, 1]), (len(df), 1)))
```

```python
# Create mini-batch data iterator
def data_iter(batch_size, features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))

    # The examples are read at random, in no particular order
    np.random.shuffle(indices)
    for i in range(0, num_examples, batch_size):
        batch_indices = np.array(
            indices[i: min(i + batch_size, num_examples)])

        yield features[batch_indices], labels[batch_indices]
```

```python
# Plot the data for Exploration
plt.scatter(features, labels)
plt.show()
```

```python
# Create Linear regresor model
lr = LinearRegression()

# Fit the linear regressor on given data
lr.fit(features, labels)
```

```python
# Predict on training data
y_pred = lr.predict(features)

# Plot the predicted outcome
plt.scatter(features, labels, marker = '.')
plt.plot(features, y_pred, color = 'orange')
plt.show()
```

```python
# Print the coefficients
print("Coefficients: \n", lr.coef_)
```

```python
# Print the mean squared error
print("Mean squared error: %.2f"%mean_squared_error(features, labels))
```

```python
# Print the coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(labels, y_pred))# The coefficients
```
