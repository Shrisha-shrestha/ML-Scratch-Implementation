```python
x1 = np.random.randn(100)
y = 2* x1 + 5 + np.random.randn(100) * 0.9

import matplotlib.pyplot as plt
plt.scatter(x1, y)
```

```python
x1 = (x1 - np.mean(x1)) / np.std(x1)
plt.scatter(x1, y)
```

```python
w=1
b=0

epochs =100
alpha = 0.1
```

```python
def simple_linear_regression(w,b,x1):
    y_hat = w * x1 + b
    return y_hat

def RMSE(y, y_hat):
    return np.sqrt(np.sum(np.square( y_hat - y)) )
```

```python
y_hat = simple_linear_regression(w, b,x1)

for i in range(epochs):
    loss = RMSE(y, y_hat)
    dw = np.sum(np.dot((y_hat - y),x1))/2
    db = np.sum(y_hat - y)/2
    w-= alpha * dw/len(x1)
    b-= alpha * db/len(x1)
    y_hat = simple_linear_regression(w, b,x1)
    print("Loss in epoch {} is {}".format(i,RMSE(y, y_hat)))
    
```

```python
def scale(d):
    mean =d.mean()
    std = d.std()
    return (d - mean) / (std)

# Create mini-batch data iterator
def data_iter(batch_size, features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))

    # The examples are read at random, in no particular order
    np.random.shuffle(indices)
    for i in range(0, num_examples, batch_size):
        batch_indices = np.array(
            indices[i: min(i + batch_size, num_examples)])

        yield features[batch_indices], labels[batch_indices]
```
