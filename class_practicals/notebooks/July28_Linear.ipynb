```python
w = 2
b = 5
x = np.random.randint(100, 200, 100)
random_noise = np.random.rand(100)
y = 2 * x + 5 + random_noise

print(y.shape)
print(y)
```

```python
x.shape, y.shape
```

```python
w1 = np.random.rand()
w0 = np.random.rand()
print(f"w1: {w1}, w0: {w0}")
```

```python
y_predicted = w1 * x + w0 
print(y_predicted)
```

```python
error = y_predicted - y
print("Error: ", np.sum(error)/100)
```

```python
learning_rate = 0.1
w1 = w1 - learning_rate * np.dot(error, x) / 100
w0 = w0 - learning_rate * np.sum(error) / 100
print(w1, w0)
```

```python
w1 = np.random.rand()
w0 = np.random.rand()

epochs = 10
learning_rate = 0.001

for t in range(epochs):
    y_predicted = w1 * x + w0

    error = y - y_predicted
    
    avg_error = np.sum(error) / len(y)

    print(f"Epoch: {t}/{epochs}, Average Error: {avg_error}")

    w1 = w1 - learning_rate * np.dot(error, x) / 100
    w0 = w0 - learning_rate * np.sum(error) / 100

print(f"after the completion of training w1 = {w1} , w0 = {w0}")
```
