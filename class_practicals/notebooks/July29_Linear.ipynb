```python
def mse(y,y_predicted):
  """
  Compute Mean Squared Error
  """
  return np.sum(( y - y_predicted)**2) / len(y)

# y = np.random.rand(2)
# y_predicted = np.random.rand(2)
# mse(y,y_predicted)
```

```python

# from evaluation import mse

class LinearRegression:
    def  __init__(self, x , y):
        self.weights = {
            "w1" : np.random.rand(),
            "w0" : np.random.rand()
        }

        self.x = x
        self.y = y


    def linear_functions(x, weights):
        """
        A linear function that process single varibale in data with bias.

        :param x: input vector.
        :param weights: dictionary of coefficient.

        :returns y: numpy array after linear transformation

        """
        w1 = weights["w1"]
        w0 = weights["w0"]

        y = w1 * x + w0

        return y

    def train(self, epochs, learning_rate):
        """
        For training linear regression using grdient descent.

        """
        for t in range(epochs):
            y_predicted = self.predict()

            mse_error = mse( self.y, y_predicted)
            print(f"Error : {mse_error}")

            error = y_predicted - self.y

            self.weights["w1"] = self.weights["w1"] - learning_rate * np.dot(error, self.x)

            print("New w1: ", self.weights["w1"])

    #self as parameter passes object to the function

    def predict(self):
        """
        A linear function that process single varibale in data with bias.

        :returns y: numpy array after linear transformation

        """
        w1 = self.weights["w1"]
        w0 = self.weights["w0"]

        y_predicted = w1 * self.x + w0

        return y_predicted

```

```python
x =  np.random.rand(10)
y = 2 * x + 7 + (np.random.rand() * 2)

EPOCHS = 10
LEARNING_RATE = 0.00001

linear_regression = LinearRegression(x, y)
linear_regression.train(epochs=EPOCHS, learning_rate=LEARNING_RATE)
```
